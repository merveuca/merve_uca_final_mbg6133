```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
sns.set_context('poster')
```

```python
random_array= np.random.randint(0,101, size=(100,100))
```

```python
random_array
```

```python
fig,ax = plt.subplots(figsize=(20,20))

sns.heatmap(random_array)
plt.show()
```
